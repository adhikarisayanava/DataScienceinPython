```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

crypto_df = pd.DataFrame(data = {'allocation %':[10, 10, 10, 60, 10]}, index=['BTC', 'ETH', 'LTC', 'XRP', 'ADA'])
print(crypto_df)
explode = (0,0,0,0.2,0)

crypto_df.plot.pie(y = 'allocation %', explode = explode ,figsize = (4, 4))
plt.title('CRYPTO PORTFOLIO PIE CHART')

```
