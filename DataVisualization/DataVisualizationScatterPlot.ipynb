```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

daily_return_df = pd.read_csv('crypto_daily_returns.csv')
print(daily_return_df)
daily_return_df.plot.scatter('BTC','ETH', grid=True, figsize=(10,5), color = 'hotpink', alpha =0.8)
plt.title('My First Scatter Plot!')
daily_return_df.plot.scatter('BTC','ADA', grid=True, figsize=(10,5), color = 'green', alpha =0.1)
plt.title('My Second Scatter Plot!')
```
