```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

daily_return_df = pd.read_csv('crypto_daily_returns.csv')
mu = round(daily_return_df['BTC'].mean(), 2)
sigma = round(daily_return_df['BTC'].std(), 2)

num_bins = 40
#plt.figure(figsize=(10,5))

daily_return_df[['BTC', 'ETH']].plot.hist(alpha = 0.7, figsize=(10,5), bins=num_bins)
plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu) + ',sigma='+ str(sigma))
```
