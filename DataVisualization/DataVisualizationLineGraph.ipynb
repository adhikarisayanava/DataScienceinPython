```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

investment_df = pd.read_csv('crypto_daily_prices.csv')
#print(investment_df)

#print(investment_df.isnull())
investment_df.plot(x = 'Date', y = ['BTC', 'ETH', 'ADA'] , subplots=True ,linewidth = 3, figsize = (14, 6))
plt.ylabel('Price [$]')
plt.title('My First Data Visualization Plot!')
plt.legend(loc = "upper right")
plt.grid()
```
